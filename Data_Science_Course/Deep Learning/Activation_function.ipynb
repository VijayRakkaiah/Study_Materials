```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# input data
x = np.linspace(-10,10,1000)
```

```python
# sigmoid
def sigmoid(x):
  return 1/(1+np.exp(-x))

plt.plot(x, sigmoid(x), label = "Sigmoid")
```

```python
def relu(x):
  return np.maximum(0,x)

plt.plot(x, relu(x), label = "ReLU")
```

```python
def leaky_relu(x):
  #return np.maximum(0.1*x, x)
  return np.where(x>0, x, 0.1*x)

plt.plot(x, leaky_relu(x), label = "Leaky ReLU")
```

```python
def softmax(x):
  #return np.exp(x)/np.sum(np.exp(x))
  exps = np.exp(x-np.max(x))# stability improve
  return exps/np.sum(exps)


plt.plot(x, softmax(x), label = "Softmax")
```

```python
def tanh(x):
  return np.tanh(x)

plt.plot(x, tanh(x), label = "Tanh")
```

```python
def swish(x):
  return x*sigmoid(x)

plt.plot(x, swish(x), label = "Swish")
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.plot(x, sigmoid(x), label = "Sigmoid", color = 'blue')
plt.title("Sigmoid")
plt.grid()

plt.subplot(2,3,2)
plt.plot(x, relu(x), label = "ReLU", color = 'green')
plt.title("Relu")
plt.grid()

plt.subplot(2,3,3)
plt.plot(x, leaky_relu(x), label = "Leaky ReLU", color = 'red')
plt.title("leaky relu")
plt.grid()

plt.subplot(2,3,4)
plt.plot(x, softmax(x), label = "Softmax", color = 'orange')
plt.title("softmax")
plt.grid()

plt.subplot(2,3,5)
plt.plot(x, tanh(x), label = "Tanh", color = 'purple')
plt.title("Tanh")
plt.grid()

plt.subplot(2,3,6)
plt.plot(x, swish(x), label = "Swish", color = 'gray')
plt.title("swish")
plt.grid()
```
