# Probability

```python
## Calculating probability of events

# Sample Space
num_cards = 52
# Favorable Outcomes
num_aces = 4 # 1 out
num_hearts = 13 # 2 out
num_diamonds = 13 # 3 out
# Divide favorable outcomes by the total number of elements in the sample space
prob_ace = num_aces / num_cards # 4/52

prob_red_card = (num_hearts+ num_diamonds) / num_cards # 26/52 # union
# Print probability rounded to two decimal places
print('The probability of getting an ace =', round(prob_ace, 3))
print('The probability of getting a red card =',round(prob_red_card, 3))

# to print the probability in percentage
prob_ace_percent = prob_ace * 100
prob_red_percent = prob_red_card * 100
print('\nThe probability of getting an ace in percentage=', str(round(prob_ace_percent, 1)) + '%')
print('The probability of getting a red card in percentage=', str(round(prob_red_percent, 1)) + '%')
```

```python
## Probability of Independent and dependent Events 
 

# Sample Space 
total_cards = 52 
cards_drawn = 1 
# remaining cards when the card drawn in the first place is not replaced 
rem_cards = total_cards - cards_drawn 

# Calculate the joint probability of drawing a king after drawing a queen in the first draw with replacement 
num_queens = 4 
num_kings = 4 
prob_king1 = num_kings / total_cards 
prob_queen1 = num_queens / total_cards 
 
# probability of intersection of events 
prob_king_and_queen1 = prob_king1 * prob_queen1 
 
# Determine the joint probability of drawing a king after drawing a queen in the first draw without replacement 
prob_king2 = num_kings / rem_cards 
prob_queen2 = num_queens / total_cards 
 
# probability of intersection of events 
prob_king_and_queen2 = prob_king2 * prob_queen2 

# Print each probability 
print('The product of probability of getting a king and the probability of drawing a queen = ',round(prob_king1 * prob_queen1,5)) 
print('The probability of getting a king after drawing a queen in the first draw with replacement = ',round(prob_king_and_queen1,5)) 
print('The probability of getting a king after drawing a queen in the first draw without replacement = ',round(prob_king_and_queen2,5)) 
```

```python
# calculating the probability as a degree of belief 
# calculating the probability of heart disease when a patient is obese 

# defining a function to find out P(A|B) given P(A), P(B), P(B|A) 
def bayes_rule(p_a, p_b, p_b_over_a): 
    # calculate P(A|B) 
    p_a_over_b = (p_b_over_a * p_a) / p_b 
    return p_a_over_b 

 
### testing of the funciton that employs Bayes rule to get posterior from prior 
p_a = 0.15 # heart dis
p_b_over_a = 0.2 
p_b= 0.1 # obese
 
# calculating the posterior P(A|B) 
p_a_over_b = bayes_rule(p_a, p_b, p_b_over_a) 
# summarize 
print('The posterior probability P(A|B) = %.1f%%' % (p_a_over_b * 100)) 
```

# Probability Mass Function (PMF)

```python
### Computing and Plotting probability mass function of a discrete random variable 

import numpy as np 
import matplotlib.pyplot as plt 
 
# inline magic function makes the plots to be stored in the notebook document 
#%matplotlib inline 
 
 ## We repeat the experiment multiple times to get a better result of probabilities 
num_rep_exp = 100 
 
# variable num_tails_rep saves the number of tails got from each run of the experiment 
num_tails_rep = np.zeros(num_rep_exp) 
 
for i in range(num_rep_exp): 
    # 
    outcome1 = np.random.choice(['H', 'T']) # specified outcome
    outcome2 = np.random.choice(['H', 'T']) 
 
    # counting the number of tails from outcome1 and outcome2 
    if outcome1 == 'H' and outcome2 == 'H': 
        num_tails = 0 
    elif outcome1 == 'H' and outcome2 == 'T': 
        num_tails = 1 
    elif outcome1 == 'T' and outcome2 == 'H': 
        num_tails = 1 
    else: # T, T
        num_tails = 2 
    
    # Saving the number of tails from each experiment at different indices of num_tails_rep. 
    num_tails_rep[i] = num_tails 
 
outcome_value, outcome_count = np.unique(num_tails_rep, return_counts=True) 
prob_count = outcome_count / len(num_tails_rep) 
 
# Now that we have tossed the coin twice for 1000 times, we plot the results 
plt.bar(outcome_value, prob_count) 
plt.ylabel('Probability') 
plt.xlabel('Outcome') 
plt.title('Probability Mass Function') 
plt.show() 
```

```python
## Binomial Random variable 
import numpy as np 
import seaborn as sns 
from scipy.stats import binom 

## n corresponds to the number of trials in each experiment, size refers to total runs of the experiment, 
## p is the probability of success. 
binom_rv = binom.rvs(n=20,p=0.5,loc=0,size=100) 
 
print('Number of successes in each trial having 20 coin tosses =', binom_rv) 

## distplot from seaborn library is used to visualize probability distributions 
ax = sns.distplot(binom_rv, color='blue', hist_kws={'linewidth': 10,'alpha':1}) 

# hist_kws specifies keywords to be used with histogram, 
# linewidth specifies the width of bars and alpha is used to adjust the color strength 
ax.set(xlabel='Values of Binomial RV', ylabel='Relative Frequency (Probability)', title ='Binomial Distribution') 
```

# cumulative frequency distribution

```python
import matplotlib.pyplot as plt 
import numpy as np 

weight_range = ['less than 50 kg', 'up to 60 kg', 'up to 70 kg', 'up to 80 kg', 'up to 90 kg', 'up to 100 kg'] 
num_students = [4, 15, 20, 22, 5, 2] 
cum_freq_students = np.cumsum(num_students) 

# plotting the frequency distribution 
plt.figure(figsize=[10,8]) 

plt.bar(weight_range, cum_freq_students) 
plt.xlabel('Range of weights',fontsize=12) 
plt.ylabel('Frequency',fontsize=12) 
plt.title('Number of students up to a specific weight', fontsize=12) 
plt.show() 
```

```python
from scipy.stats import norm 
import numpy as np 
import matplotlib.pyplot as plt 
 
# Generating a range of values from -4 to 4 because a standard Normal RV has most values between -3 to 3 
x= np.arange(-4,4,0.01) 
 
# Plot of PDF 
plt.plot(x,norm.pdf(x)) 
plt.xlabel('Values of RV X') 
plt.ylabel('Probability') 
plt.title('Probability Density Function of a Normal RV') 
plt.show() 
```

```python
# Plot of CDF 
plt.plot(x,norm.cdf(x)) 
plt.xlabel('Values of RV X') 
plt.ylabel('Probability') 
plt.title('Cumulative Distribution Function of a Normal RV') 
plt.show() 
```

```python

# Plot of Inverse CDF (or PPF or Quantile function) 
plt.plot(x, norm.ppf(x)) 
plt.xlabel('Probability') 
plt.ylabel('Values of RV X') 
plt.title('Quantile Function of a Normal RV') 
plt.show() 
```

# Statistical Inference

```python
# Point Estimation
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import scipy.stats as stats 
 
np.random.seed(12345) 
population = stats.poisson.rvs(loc=12, mu=24, size=10000) 
print('The population mean is', population.mean()) 
```

```python
pd.Series(population).hist(density=True, zorder=1, alpha=1, label='Population Histogram')

# We estimate mean by taking a sample out of original population 
np.random.seed(12345) 
sample = np.random.choice(a= population, size=500) # Sample 500 values 
print ('The sample mean is', sample.mean()) 

pd.Series(sample).hist(density=True, zorder=2, alpha=0.5, label='Sample Histogram')

plt.legend() 
 
# We check the difference between two means 
print('Difference of the means is ' ,population.mean() - sample.mean() ) # difference between means 
```
