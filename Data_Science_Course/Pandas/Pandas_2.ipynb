```python
data_2 = pd.read_csv('gun-violence-data1.csv')
#data_2 = pd.read_csv('C:/Users/AJANTHA/Downloads/Python_Introduction/Guvi_2025/gun-violence-data1.csv')
```

```python
data_2.head()
```

```python
from datetime import datetime
```

```python
#data_2['date'] = pd.to_datetime(data_2['date'])
data_2['date'] = pd.to_datetime(data_2['date'],dayfirst=True)
#data_1['date'] = pd.to_datetime(data_1['date'],format = "%d-%m-%Y")
```

```python
data_2.head()
```

```python
data_2['day'] = data_2['date'].dt.day
data_2['month']=data_2['date'].dt.month # extract month
data_2['year']=data_2['date'].dt.year # extract year
data_2['day_of_week']=data_2['date'].dt.dayofweek 
data_2['tot_victim']  = data_2['n_killed']+data_2['n_injured']
data_2.head()
```

```python
data_2['day_of_week'] = data_2['day_of_week'].map({0:'Monday',1:"Tue", 2:'wed',3:"thur",4:'fri',5:'sat',6:'sun'})
data_2.head()
```

```python
# grouping and sorting
data_2.shape
```

```python
data_2['day_of_week'].value_counts()
```

```python
data_2.groupby('day_of_week')['tot_victim'].count()
```

```python
data_2.groupby('day_of_week')['n_killed'].count()
```

```python
data_2.groupby('state')['tot_victim'].sum()
```

```python
import numpy as np
data_2.groupby('month')['tot_victim'].agg([np.mean,max])
```

```python
# sorting
data_2.groupby('state')['tot_victim'].sum().sort_values(ascending = False)
```

```python
data_2.groupby('month')['tot_victim'].count().sort_values(ascending = True)
```
