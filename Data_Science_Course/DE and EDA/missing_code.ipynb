```python
df = pd.read_csv('loan-prediction.csv')
df.head()
```

```python
df.describe()
```

```python
# delete row
df1 = df.dropna(axis = 0)
df1.isnull().sum()
```

```python
df1.info()
```

```python
# delete col
df2 = df.drop(['Self_Employed'], axis = 1)
df2.isnull().sum()
```

```python
df.info()
```

```python
df2.isnull().sum()
```

```python
# impute => replace => specific => 0
df2['LoanAmount'] = df2['LoanAmount'].fillna(0)
df2.isnull().sum()
```

```python
df2.describe()
```

```python
df2.head(20)
```

```python
# impute => mean
df2['Credit_History'] = df2['Credit_History'].fillna(df2['Credit_History'].mean())
```

```python
df2.head(20)
```

```python
# impute => mode
df2['Gender'] = df2['Gender'].fillna(df2['Gender'].mode()[0])
df2['Married'] = df2['Married'].fillna(df2['Married'].mode()[0])
df2.isnull().sum()
```

```python
df2['Loan_Amount_Term'] = df2['Loan_Amount_Term'].fillna(df2['Loan_Amount_Term'].median())
df2.isnull().sum()
```

```python
# categorical

x = pd.DataFrame({'Shape':['sq','cir','sq','oval',np.nan]})
x
```

```python
from sklearn.impute import SimpleImputer
impute = SimpleImputer(strategy='most_frequent')
impute.fit_transform(x)
```

```python
impute = SimpleImputer(strategy='constant', fill_value='missing')
impute.fit_transform(x)
```

```python

imp= SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit([[1,2],[np.nan,3],[7,6]])
```

```python
x = [[1,2],[np.nan,3],[7,6]]
print(imp.transform(x))
```

```python
x1 = [[np.nan, 2],
     [6,np.nan],
     [7,6]]
print(imp.transform(x1))
```
