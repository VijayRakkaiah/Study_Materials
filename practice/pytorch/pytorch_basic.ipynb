```python
# step 1
# Neural Net
import torch.nn as nn
```

```python
# Step 2 --> Define
# define input layer, hidden layes, output layer and batch
n_in, n_h, n_out, batch_size = 10, 15, 1, 10
```

```python
# Step 3 --> Combination
x = torch.randn(batch_size, n_in)
y = torch.tensor([[1.0], [0.0], [0.0], [1.0], [1.0], [1.0], [0.0], [0.0], [1.0], [1.0]])
```

```python
# Step 4 --> Model creation
# Create a model

model = nn.Sequential(nn.Linear(n_in, n_h),
                      nn.ReLU(),
                      nn.Linear(n_h, n_out),
                      nn.Sigmoid())
```

```python
# Step 5 --> Loss Function
# Gradient Descent
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) #Stochastic Descent
```

```python
# Step 6
for epoch in range(15):
  #fwd pass
  y_pred = model(x)

  # loss
  loss = criterion(y_pred, y)
  print('epoch: ', epoch, 'loss: ', loss.item())

  # Backward Pass
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 1. Define model
class SimpleNet(nn.Module):
  def __init__(self):
    super(SimpleNet, self).__init__()
    self.fc = nn.Linear(10, 1)

  def forward(self, x):
    return self.fc(x)

model = SimpleNet()

# 2. Define loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# 3. Training data
x = torch.randn(100, 10)
y = torch.randn(100, 1)

# 4. Training loop
for epoch in range(10):
  outputs = model(x)
  loss = criterion(outputs, y)

  optimizer.zero_grad() # Clear previous gradient
  loss.backward()       # Backprobagation
  optimizer.step()      # Update Weights

  print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```
